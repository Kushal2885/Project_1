2. Explain the difference between:
A) Mean, Median, Mode
Measure	Definition	Formula	Example
    Mean	Average value of data	Sum of values / Total values	Average income
    Median	Middle value after arranging data	Middle observation	Middle family income
    Mode	Most frequently occurring value	Highest frequency	Most common education level
Example:

    Dataset = 10, 20, 20, 30, 40

    Mean = (10+20+20+30+40)/5 = 24
    Median = 20
    Mode = 20
B) Range, Variance, Standard Deviation
    Measure	Definition	Formula	Purpose
    Range	Difference between maximum and minimum value	Max − Min	Measures spread
    Variance	Average squared deviation from mean	σ²	Measures variability
    Standard Deviation	Square root of variance	√Variance	Measures data dispersion
Example:

    Dataset = 5, 10, 15

    Range = 15 − 5 = 10
    Mean = 10
    Variance = [(5−10)² + (10−10)² + (15−10)²]/3 = 16.67
    Standard Deviation = √16.67 = 4.08

    3. Important Statistical Terms
A) Gaussian Distribution (Normal Distribution)

Gaussian Distribution is a symmetric bell-shaped distribution where most values are concentrated around the mean.

Characteristics:
    Mean = Median = Mode
    Symmetrical curve
    Follows bell shape

B) Log Normal Distribution

A Log Normal Distribution occurs when the logarithm of data is normally distributed.

It is positively skewed.

Characteristics:
    Cannot contain negative values
    Used for income, stock prices, population growth

C) 3-Sigma Rule (Empirical Rule)

The 3-Sigma Rule states that in a normal distribution:

    68% data lies within 1 standard deviation
    95% data lies within 2 standard deviations
    99.7% data lies within 3 standard deviations

D) Percentiles

Percentiles divide data into 100 equal parts.

Common Percentiles:
    25th Percentile
    50th Percentile (Median)
    75th Percentile

E) Quartiles

Quartiles divide data into 4 equal parts.

Quartile
    Q1	First 25% data
    Q2	Median (50%)
    Q3	75% data

F) Five Number Summary

Five Number Summary describes dataset distribution using five important values:

    1. Minimum
    2. Q1
    3. Median
    4. Q3
    5. Maximum

G) Skewness

Skewness measures asymmetry in data distribution.

Types:
    1. Positive Skew
    2. Negative Skew
    3. Zero Skew

H) Kurtosis

Kurtosis measures the peakedness or flatness of distribution.

Types:
    1. Leptokurtic → High peak
    2. Mesokurtic → Normal peak
    3. Platykurtic → Flat peak


```python
#Part B: Practical Implementation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("household_socioeconomic_dataset.csv")

df.info()
```

```python
#3. Types of Data

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

```python
#4. Central Tendency

income_mean = df['Household_Income'].mean()
age_mean = df['Age_of_Household_Head'].mean()

income_median = df['Household_Income'].median()
age_median = df['Age_of_Household_Head'].median()

income_mode = df['Household_Income'].mode()[0]
age_mode = df['Age_of_Household_Head'].mode()[0]

print("Income Mean:", income_mean)
print("Income Median:", income_median)
print("Income Mode:", income_mode)

print("\nAge Mean:", age_mean)
print("Age Median:", age_median)
print("Age Mode:", age_mode)
```

```python
#5. Measures of Dispersion

income_range = df['Household_Income'].max() - df['Household_Income'].min()
income_variance = df['Household_Income'].var()
income_std = df['Household_Income'].std()
print("Range:", income_range)
print("Variance:", income_variance)
print("Standard Deviation:", income_std)


Q1 = df['Household_Income'].quantile(0.25)
Q3 = df['Household_Income'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
#6. Distribution Analysis

plt.figure(figsize=(10,6))
plt.hist(df['Household_Income'], bins=15)
plt.title("Histogram of Household Income")
plt.xlabel("Household Income")
plt.ylabel("Frequency")

plt.show()
```

```python
# Gaussian Normal Distribution Curve

from scipy.stats import norm

plt.figure(figsize=(10,6))

plt.hist(df['Household_Income'], bins=15, density=True)

mu, std = norm.fit(df['Household_Income'])

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

p = norm.pdf(x, mu, std)

plt.plot(x, p, linewidth=2)

plt.title("Gaussian Normal Distribution Curve")
plt.xlabel("Income")
plt.ylabel("Density")

plt.show()

print("Mean:", mu)
print("Standard Deviation:", std)


skewness = df['Household_Income'].skew()
kurtosis = df['Household_Income'].kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
#7. Data Categorization

plt.figure(figsize=(10,6))

df.boxplot(column='Household_Income', by='Education_Level')

plt.title("Income by Education Level")


plt.xlabel("Education Level")
plt.ylabel("Income")

plt.show()
```
